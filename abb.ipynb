```python
import altair as alt
import pandas as pd

df = pd.read_csv("listings.csv")

df.head()

listings = df[['neighbourhood_cleansed','estimated_occupancy_l365d','reviews_per_month', 'price', 'number_of_reviews', 'availability_365','property_type','price_quote_price_per_night','estimated_revenue_l365d','bathrooms','bedrooms','beds','minimum_nights','maximum_nights','review_scores_rating']]
listings.head()

listings = listings.dropna()
listings = listings[listings['number_of_reviews'] > 0]

listings.head()
```

```python
alt.Chart(listings).mark_rect().encode(
    x = alt.X('minimum_nights:Q', bin = alt.Bin(maxbins = 20)),
    y = alt.Y('estimated_occupancy_l365d:Q', bin = alt.Bin(maxbins = 20)),
    color = "count():Q"
)
```

```python

listings = listings[listings['price_quote_price_per_night'] < 2000]
#remove approx 5 outliers for estimated revenue
alt.Chart(listings).mark_bar().encode(
    x = alt.X("bedrooms:Q"),
    y = alt.Y("price_quote_price_per_night:Q", aggregate = "mean"),
    color = "count():Q"
)
```

```python
#listings = listings.groupby(['property_type']).count().nlargest(5, 'property_type')
alt.Chart(listings).mark_circle().encode(
    x = alt.X('price_quote_price_per_night:Q'),
    y = alt.Y('estimated_revenue_l365d:Q'),
    color = "property_type:N",
)
```

```python
listings_top_property = listings.groupby('property_type')['estimated_revenue_l365d'].mean().nlargest(5).reset_index()
#print(listings_top_property)
top_property_types = listings_top_property['property_type'].tolist()
listings_top_types = listings[listings['property_type'].isin(top_property_types)]

alt.Chart(listings_top_types).mark_circle().encode(
    x = alt.X('price_quote_price_per_night:Q'),
    y = alt.Y('estimated_revenue_l365d:Q'),
    color = "property_type:N",
).properties(
    title = ("Estimated Revenue VS Price Per Night For The Top 5 Property Types")
)
```
